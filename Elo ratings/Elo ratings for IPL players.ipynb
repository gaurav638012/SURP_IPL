```python
import pandas as pd
import numpy as np

ratings = {}
init = 1000
diff = 400
chances = 10
kfac = 30
```

```python
file = open("clean.csv", 'r')

for line in file:
    l = line.split(",")
    if l[0] == 'id': continue
    
    player1 = l[4]
    player2 = l[6]
    
    if player1 not in ratings.keys(): ratings[player1] = init
    if player2 not in ratings.keys(): ratings[player2] = init
        
    prob = 1/(1+pow(chances, (ratings[player2]-ratings[player1])/diff))
    
    batsman_runs = int(l[7])
    if l[15] != "legbyes" or l[15] != "byes": bowler_runs = int(l[9])
    else: bowler_runs = 0
    if l[13] == player1: batsman_wicket = 1
    else: batsman_wicket = 0
    if l[11] == "1" and l[12] != "run out": bowler_wicket = 1
    else: bowler_wicket = 0
     
    batsman_outcome = (batsman_runs+30*(1-batsman_wicket))/36
    bowler_outcome = (bowler_runs+30*(1-bowler_wicket))/37
    
    ratings[player1] += kfac*(batsman_outcome-prob)
    ratings[player2] -= kfac*(bowler_outcome-prob)
```

```python
for w in sorted(ratings, key=ratings.get, reverse=True):
    print(w, ratings[w])
```
